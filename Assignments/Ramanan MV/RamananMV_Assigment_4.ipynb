# Find the outliers and replace them outliers

```python
def find_outliers(df):
   q1=df.quantile(0.25)
   q3=df.quantile(0.75)
   inter_quantile_distance=q3-q1
   outliers = df[((df<(q1-1.5*inter_quantile_distance)) | (df>(q3+1.5*inter_quantile_distance)))]
   return outliers
```

```python
outliers = find_outliers(df["Viscera weight"])
print("No of outliers: " + str(len(outliers)))
print("Maximum outlier value: " + str(outliers.max()))
print("Minimum outlier value: " + str(outliers.min()))
print()
outliers
```
